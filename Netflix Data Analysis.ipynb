```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.DataFrame(pd.read_csv('/content/drive/MyDrive/Scaler/netflix.csv'))
data.head()
```

Find out the missing or NaN values in the dataset

```python
plt.figure(figsize=(10,8))
sns.heatmap(data.isna())
```

As we can see, a lot of records in the data doesn't have values for **Director, Cast and Country** fields


Since we are interested in increasing the revenue of Netflix, our main objective is to figure out which all shows and movies performed the best.

We will firstly look at the distribution of how many tv shows and movies are present in netflix. Later on, we will specifically start analysing more details about the ones that fetched the best reviews

```python
sns.countplot(x= 'type', data=data)
```

```python
# plt.figure(figsize=(22,10))
data["listed_in"].value_counts()[:10].plot(kind="barh", color="red")
plt.title("Top 10 Genres of NETFLIX",size=20);
```

We have a lot of movies compared to TV shows, hence we will split our data so that our analysis is not skewed

Before that, we need to clean/preprocess our data to get better insights

# Movies

```python
df = data.copy()
df = df[df['type'] == 'Movie']

df.head()
```

Since we don't have much values in the **Director** and **Cast** columns, we will drop them

However we cannot drop the countries as that is one of our primary attribute for this case study

```python
df.drop(['director', 'cast'], axis=1, inplace=True)
df.head()
```

```python
df[df['country'].isnull()]
```

Assuming the data might have been collected with some sequence, we will probably start imputing with forward fill method instead of using mode or mean values to impute the data.

***Note***: **We can train a model just to impute the values, but that will consume a lot of our time hence parked for later improvements.**

```python
df['country'] = df['country'].ffill(axis=0)
df[df['country'].isnull()]
```

```python
# Let us now also check for null rated shows/movies

df[df['rating'].isnull()]
```

Since there are only 2 of them, we can manually fill there values from internet

```python
ratings = ['TV-PG', 'TV-MA']

for id, rating in zip(df[df['rating'].isnull()].index, ratings):
  df['rating'].loc[id] = rating
```

```python
df[df['rating'].isnull()]
```

For records, where date was null, we will drop them since they are quite scarse and won't affect our analysis much.

We will look into the same later on by imputing the values manually as they are also not much in count

Let us verify again whether there is a null data or not

```python
df = df[df['date_added'].notna()]
df.info()
```

As we can see, the country column holds multiple values of countries where the content was displayed. 
Assuming the first country will be the major source of the content we are going to use those for our further analysis

```python
df['main_country'] = df['country'].apply(lambda x: x.split(',')[0])
df
```

Let us start our analysis for problem statement.

We will divide the data based on the top 10 main countries we found and group the data based on their genres and ratings.

Since we will have insights into what kind of content fetched most good reviews in which country based on top genres, we can promote the development of similar content in future

```python
# For ease of analysis, let's pick top 10 countries with maximum content produced

top_countries = df.groupby('main_country').count().sort_values('type', ascending=False)[:10]
top_countries
```

```python
# For ease of analysis, let's pick top 10 genres with maximum content produced

top_genres = df.groupby('listed_in').count().sort_values('type', ascending=False)[:10]
top_genres
```

```python
# Filtering the data for top 10 genres 

df = df[df['listed_in'].isin(list(top_genres.index))]

# Filtering the data for top 10 countries

df = df[df['main_country'].isin(list(top_countries.index))]
```

```python
df
```

```python
ratings=[]
for rate in df['rating'].unique():
  ratings.append(rate)

countries = df['main_country'].unique()

listing = df['listed_in'].unique()

ratings, countries, listing
```

**Top content based on genres and the top ratings within each in different countries is displayed below**

```python
fig = plt.figure(
    figsize=(30,40)
    )

for i, name in enumerate(countries):
  frame = df[df['main_country'] == str(name)]
  ax = fig.add_subplot(len(countries),1,i+1)
  topic = name
  sns.countplot(x='listed_in', data= frame[frame['listed_in'].isin(listing)], hue='rating')
  ax.set_title(topic)
  plt.subplots_adjust(left=0.1,
                    bottom=0.1, 
                    right=0.9, 
                    top=1.5, 
                    wspace=0.5, 
                    hspace=2.0)
  plt.xlabel('Genre')
  plt.xticks(rotation = 50)
  ax.set(ylabel='Content Produced')
```

```python
fig = plt.figure(
    figsize=(20,32)
    )

for i, name in enumerate(countries):
  frame = df[df['main_country'] == str(name)]
  ax = fig.add_subplot(len(countries),1,i+1)
  topic = name
  sns.countplot(x='rating', data= frame[frame['rating'].isin(ratings)])
  ax.set_title(topic)
  plt.subplots_adjust(left=0.1,
                    bottom=0.1, 
                    right=0.9, 
                    top=0.9, 
                    wspace=0.4, 
                    hspace=0.4)
  ax.set(ylabel='Content Produced')
```

From the above graph, we can clearly visualize what type of content and ratings are attracting the audience in top 10 countries.

Note: Top 10 genres were considered for displaying the above data
